```python
%%sql
SELECT 
    e.first_name AS name, e.gender, e.hire_date, d.dept_name
FROM
    employees AS e
        JOIN
    dept_emp de ON e.emp_id = de.emp_id
        JOIN
    departments d ON d.dept_id = de.dept_id
WHERE
    d.dept_id = 1;
```

```python
%%sql
SELECT 
    COUNT(emp_id), salary
FROM
    salaries
GROUP BY salary;
```

```python
%%sql
SELECT 
    COUNT(e.emp_id), d.dept_id, d.dept_name
FROM
    employees e
        LEFT JOIN
    dept_emp de ON de.emp_id = e.emp_id
        LEFT JOIN
    departments d ON d.dept_id = de.dept_id
GROUP BY d.dept_name;
```

```python
%%sql
select count(emp_id),salary
from salaries
group by salary
having salary > 300000 ;
```

```python
%%sql
SELECT 
    COUNT(e.emp_id), d.dept_id, d.dept_name
FROM
    employees e
        LEFT JOIN
    dept_emp de ON de.emp_id = e.emp_id
        LEFT JOIN
    departments d ON d.dept_id = de.dept_id
GROUP BY d.dept_name
HAVING COUNT(e.emp_id) > 10;
```
